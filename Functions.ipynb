When you call a function, Python must match each argument in the function
call with a parameter in the function definition. The simplest way to
do this is based on the order of the arguments provided. Values matched
up this way are called **positional arguments**.

```python
def describe_pet(pet_name, pet_type):
  print('\n')
  print(f"I have a {pet_type} and its name is {pet_name}")

describe_pet('harry','horse')
```

A **keyword argument** is a name-value pair that you pass to a function. You
directly associate the name and the value within the argument, so when you
pass the argument to the function, theres no confusion

```python
describe_pet(pet_name='ross', pet_type='duck')
```

```python
#default values
def describe_pet(animal_type='dog', animal_name):
  print(f"I have a {animal_type} whose name is {animal_name}")
#When you use default values, any parameter with a default value needs to be listed
# after all the parameters that dont have default values. This allows Python to continue
# interpreting positional arguments correctly.
```

```python
def describe_pet(animal_name, animal_type='dog'):
  print(f"I have a {animal_type} whose name is {animal_name}")

describe_pet('George')
```

Making an argument optional

```python
# set the default value to '' or None and then define condition statements on it
def name_formatted(first_name, last_name, middle_name=''):
  if middle_name:
    print(middle_name)
  else:
    print('zilch')
name_formatted('jk', 'lk')
```

```python
def name_formatted(first_name, last_name, middle_name=''):
  if middle_name:
    return(f" {first_name.title()} {middle_name.title()} {last_name.title()}")
  else:
    return(f" {first_name.title()} {last_name.title()}")
```

```python
name_formatted('Mike', 'Kazz')
```

**Passing an arbitrary number of arguments**
*args

```python
def make_pizza(*toppings):
  print(toppings)
make_pizza('mushroom')
make_pizza('pepperoni', 'pine apple', 'cheese')
#the functions packs the output into a tuple; even if it is only one
```

**Passing an arbitrary number of keyword arguments**
**kwargs

```python
def user_build(first_name, last_name, **user_info):
  #python builds a dictionary for kwargs
  print(first_name.title(), last_name.title())
  print(user_info)

user_build('Mike', 'Kazz', age=24, height=185)
```

```python
def user_build(first_name, last_name, **user_info):
  #python builds a dictionary for kwargs
  user_info['first name'] = first_name.title()
  user_info['last name'] = last_name.title()
  print(user_info)

user_build('Mike', 'Kazz', age=24, height=185)
```
