```python
# defining the class Dog
class Dog():
  def __init__(self, name, age):
    self.name = name
    self.age = age
  def sit(self):
    print(f"{self.name} is sitting down peacefully!")
  def roll_over(self):
    print(f"{self.name} is rolling over.")
```

**no parentheses means that we are building this class from the scratch.**

**if there was a parentheses it would mean that we are using inheritance.**

```python
my_doggy = Dog('sagy',12)
print(f"my_doggy's name is {my_doggy.name} and he is {my_doggy.age} years old.")
my_doggy.sit()
my_doggy.roll_over()
```

```python
class Car:
  def __init__(self, make, model, year):
    self.make = make
    self.model = model
    self.year = year
    self.odeometer_reading = 0
  def get_descriptive_name(self):
    """return a neatly formatted desctiption of the car."""
    long_name = f"{self.make} {self.model} {self.year}"
    return(long_name.title())
  def read_odeometer(self):
    print(f"this car has {self.odeometer_reading} miles on it.")
  def update_odometer(self,milage):
    if milage >= self.odeometer_reading:
      self.odeometer_reading = milage
    else:
      print('listen! you cant undo the extra milage!')
    print(f"your new milage on this car is {self.odeometer_reading}")
```

```python
my_car = Car('audi','a4', 1995)
my_car.get_descriptive_name()
my_car.read_odeometer()
```

```python
class Restaurant:
  def __init__(self,name,cuisine_type,number_served=0):
    self.name = name
    self.cuisine_type = cuisine_type
    self.number_served = number_served
  def describe_restaurant(self):
    print(self.name,self.cuisine_type)
  def open_restaurant(self):
    print(f"{self.name} restaurant is now open. welcome!")
  def customers_served(self):
    print(f"this restaurant has served {self.number_served} customers.")
  def set_number_served(self, updated_number):
    self.number_served = updated_number
  def increment_number_served(self,incrm):
    self.number_served += incrm 
    print(f"this place has served {self.number_served} customers today.")
```

```python
res3 =Restaurant('Zapata','Asian',5)
res3.customers_served()
res3.number_served = 40
res3.customers_served()
```

```python
res4 = Restaurant('Starbucks','coffee shop', 13)
res4.set_number_served(32)
res4.customers_served()
```

```python
res4.increment_number_served(9)
```

```python
class ElectricCar(Car):
  def __init__(self, make, model, year):
    super().__init__(make, model, year)
    self.battery_size = 75
  def describe_battery_size(self):
    print(f"this EV has a battery size of {self.battery_size}")
```

```python
my_tesla = ElectricCar('tesla','S', 2022)
my_tesla.get_descriptive_name()
```

```python
my_tesla.describe_battery_size()
```

```python
class Battery:
  def __init__(self,battery_size=75):
    self.battery_size = battery_size
  def describe_battery(self):
    print(f"this car has a {self.battery_size}-kWh battery.")
  def get_range(self):
    if self.battery_size == 75:
      range = 260
    if self.battery_size == 100:
      range = 315
    print(f"this car can go about {range} miles on a full charge.")
  def upgrade_battery(self):
    if self.battery_size != 100:
      self.battery_size = 100
```

```python
class ElectricCar(Car):
  def __init__(self, make, model, year):
    super().__init__(make, model, year)
    self.battery = Battery()
```

```python
EV1 = ElectricCar('BMW','electic', 2022)
EV1.battery.get_range()
EV1.battery.upgrade_battery()
EV1.battery.get_range()
```

```python
my_tesla = ElectricCar('tesla','S', 2022)
my_tesla.battery.describe_battery()
```

```python
my_tesla.battery.battery_size = 100
my_tesla.battery.describe_battery()
```

```python
my_tesla.battery.get_range()
```

```python
from random import randint
class Privileges:
  def __init__(self,privileges=set()):
    self.privileges = privileges
  def add_privileges(self):
    n = randint(1,3)
    dic ={3:'can add post', 2:'can delete post', 1:'can ban user'}
    self.privileges.add(dic[n])
  def show_privileges(self):
    print(f"this admin can {self.privileges}")
```

```python
class Admin(User):
  def __init__(self,first_name,last_name,last_visit,likes,sex,login_attempts):
    super().__init__(first_name,last_name,last_visit,likes,sex,login_attempts)
    self.privileges = Privileges()
```

##**Lottery: Make a list or tuple containing a series of 10 numbers and five letters. Randomly select four numbers or letters from the list and print a message saying that any ticket matching these four numbers or letters wins a prize.**##

```python
from random import sample, randint
lottery_list = [2,25,23,21,28,9,14,15,16,17,'a','e','w','t','y']

numbers = []
strings = []

for i in lottery_list:
  if type(i) == str:
    strings.append(i)
  else:
    numbers.append(i)

if randint(0, 1) == 0:
  print(sample(numbers, 4))
else:
  print(sample(strings, 4))
```
